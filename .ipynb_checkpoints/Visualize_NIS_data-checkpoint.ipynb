```python
import matplotlib.pyplot as plt
from pylab import *
```

```python
nisrad_lst = []
nislid_lst = []
# Read the NIS Laser input file
with open("build/NIS_Laser_output.txt") as file:
    for line in file: 
        line = line.strip()
        nislid_lst.append(float(line))
# Read the NIS Radar input file        
with open("build/NIS_radar_output.txt") as file:
    for line in file: 
        line = line.strip()
        nisrad_lst.append(float(line))
```

### Plot the NIS values for radar

```python
plt.figure(figsize=(8,4))
plt.plot(nisrad_lst)
plt.ylabel('NIS Value')
plt.xlabel('Time Step')
plt.title('Radar NIS Values')
plt.axhline(y=7.8, color='r', linestyle='-')
plt.ylim([0,15])
plt.savefig('img/NIS_Radar.png')
plt.show()
```

#### Output seems to be reasonable within 7.8 value for radar

### Plot the NIS values for Lidar

```python
plt.figure(figsize=(8,4))
plt.plot(nislid_lst)
plt.ylabel('NIS Value')
plt.xlabel('Time Step')
plt.title('Laser NIS Values')
plt.axhline(y=7.8, color='r', linestyle='-')
plt.ylim([0,15])
plt.savefig('img/NIS_Laser.png')
plt.show()
```

#### Output for laser also seems to be Ok
